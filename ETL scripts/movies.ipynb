Extraer el documento

```python
import pandas as pd
import os as os
```

```python
file_movies = 'Resources/IMDb_movies.csv'
movies_df = pd.read_csv(file_movies)
```

```python
movies_df = movies_df[['imdb_title_id','original_title','year','genre','country','director',
'production_company','actors','budget']]
movies_df.head()
```

```python
list(movies_df)
```

Extraer y poner en la lista los generos

```python
genre_df = pd.DataFrame(movies_df.genre.str.split(",",expand=True))
genre_df.head()
```

```python
genre_list = []
for i in range(0,len(genre_df)):
    for j in range(3):
        genre_list.append(genre_df.iloc[i][j])

genre = pd.DataFrame(genre_list)
genre.head()
```

```python
genre = genre.rename(columns={0:'genre'})
genre = genre['genre'].str.strip()
genre = genre.drop_duplicates()
genre = pd.DataFrame(genre)
genre
```

Extraer y hacer catalogo el country

```python
country_df = pd.DataFrame(movies_df.country.str.split(",",expand=True))
country_df.head()
```

```python
country_list = []
for i in range(0,len(country_df)):
    for j in range(24):
        country_list.append(country_df.iloc[i][j])

country = pd.DataFrame(country_list)
country.head()
```

```python
country = country.rename(columns={0:'country'})
country = country['country'].str.strip()
country = country.drop_duplicates()
country = pd.DataFrame(country)
country
```

```python
company = pd.DataFrame(movies_df['production_company'])
company = company.dropna()
company = company['production_company'].drop_duplicates()
company
```

Extraer y generar catalogo de directores

```python
director_df = pd.DataFrame(movies_df.director.str.split(",",expand=True))
director_df
```

```python
director_list = []
for i in range(0,len(director_df)):
    for j in range(1):
        director_list.append(director_df.iloc[i][j])

director = pd.DataFrame(director_list)
director.head()
```

```python
director = director.rename(columns={0:'director'})
director = director['director'].str.strip()
director = director.drop_duplicates()
director = pd.DataFrame(director)
director
```

Extraer y hacer el catalogo de actores

```python
actors_df = pd.DataFrame(movies_df.actors.str.split(',',expand=True))
actors_df.head()
```

```python
actors_list = []
for i in range(0,len(actors_df)):
    for j in range(15):
        actors_list.append(actors_df.iloc[i][j])

actors = pd.DataFrame(actors_list)
actors.head()
```

```python
actors = actors.rename(columns={0:'actors'})
actors = actors['actors'].str.strip()
actors = actors.drop_duplicates()
actors = pd.DataFrame(actors)
actors
```
